```python
# To read the csv in pandas
# since we don't need the first three unwanted rows, we can skip it using skiprows
read_csv = pd.read_csv("test_input_extract_csv.csv",skiprows=3)  
#once the rows are skipped, we need to store in another file 
modified = read_csv.to_csv("test_output_modified_csv.csv")
```

```python
#Read the csv to check we have skipped the rows
read = pd.read_csv("test_output_modified_csv.csv",index_col=0)
read
```

```python
# We need to format the column Txn Date, Value Date and Bookind Date in yyyy/mm/dd format and replace the integers' commas
read["Txn Date"] = pd.to_datetime(read["Txn Date"]).dt.strftime("%Y/%m/%d")
read["Value Date"] = pd.to_datetime(read["Value Date"]).dt.strftime("%Y/%m/%d")
read["Booking Date"] = pd.to_datetime(read["Booking Date"],errors='coerce').dt.strftime("%Y/%m/%d")
read["Debit"]=read["Debit"].str.replace(",","").astype(float)
read["Credit"]=read["Credit"].str.replace(",","").astype(float)
read["Balance"]=read["Balance"].str.replace(",","").astype(float)
```

```python
# Once we have made corrections using the above code 
# store the csv files and excel files with same format
Export_Properly_Modified = read.to_csv("test_output_modified_csv.csv")
Export_Properly_Modified_excel = read.to_excel("test_output_modified_csv.xls")
```
